```python
import requests
import xmltodict
import yaml
```

```python
original_text = "ぼくらの7日間戦争"
original_text = "title exact " + original_text
encoded_text = original_text.encode('utf-8')
print(encoded_text)
```

```python
params = {
    "operation": "searchRetrieve",
    'query': encoded_text,
    'recordSchema': 'dcndl',
    'maximumRecords': 10,
    "onlyBib": "true",
}

response = requests.get('https://ndlsearch.ndl.go.jp/api/sru', params=params)
print(response.text)
```

```python
dict_data = xmltodict.parse(response.text)
yaml_data = yaml.dump(dict_data["searchRetrieveResponse"]["records"], allow_unicode=True)
with open('data.yaml', 'w') as f:
    f.write(yaml_data)
```
